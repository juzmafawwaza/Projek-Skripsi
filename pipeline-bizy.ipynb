# Answer Classification

Proses ini melibatkan pembuangan data Human dan membuat timestamp dari informasi created_id

```python
compile = compile[compile["type"]=="ai"]
compile = compile.drop(['additional_kwargs','example'], axis=1)
compile = compile.rename(columns={'created_at': 'time'})
series_datetime = pd.to_datetime(compile['time'])
compile['created_at'] = series_datetime.dt.strftime("%m/%d/%Y")
compile['timestamp'] = compile['time'].str[-14:]
first_column = compile.pop('timestamp')
compile.insert(0, 'timestamp', first_column)
first_column = compile.pop('created_at')
compile.insert(0, 'created_at', first_column)
compile = compile.drop(['time'],axis = 1)

compile
```

```python
keywords1 = ["maaf", "tidak", "informasi"]
keywords2 = ["maaf", "belum", "informasi"]
keywords3 = ["sorry"]
keywords4 = ["maaf","tidak","mengerti"]
keywords5 = ["maaf","tidak tahu"]

condition1 = (
    compile['content'].str.contains(keywords1[0], case=False, na=False) &
    compile['content'].str.contains(keywords1[1], case=False, na=False) &
    compile['content'].str.contains(keywords1[2], case=False, na=False)
)

condition2 = (
    compile['content'].str.contains(keywords2[0], case=False, na=False) &
    compile['content'].str.contains(keywords2[1], case=False, na=False) &
    compile['content'].str.contains(keywords2[2], case=False, na=False)
)

compile['Ketepatan Menjawab'] = (condition1 | condition2).astype(int)

print(compile)
```

```python
compile = compile.reset_index(drop=True)
```

# Preprocessing Data

```python
import pandas as pd

input_data = pd.read_csv('/kaggle/working/Compiled.csv')
input_data = input_data[input_data["Ketepatan Menjawab"] == 1]
input_data = input_data.reset_index(drop=True)
len(input_data)
```

```python
input_data['prepro_text'] = input_data['content']
print(input_data)
```

```python
import pandas as pd
import re

# Fungsi baru untuk mengekstrak kalimat penting
def extract_important_sentence(content):
    # Ekstrak kalimat yang dimulai dengan variasi kata 'Maaf'
    match = re.search(r'((?:saya mohon maaf|mohon maaf|maaf)[^.]*)\.', content, re.IGNORECASE)
    extracted_sentence = match.group(0) if match else None
    return extracted_sentence

# Terapkan fungsi ke DataFrame untuk mengekstrak kalimat penting
input_data['prepro_text'] = input_data['prepro_text'].apply(extract_important_sentence)

# Lihat hasilnya
input_data
```

# Intent Classification

```python
from transformers import BertTokenizer
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import load_model
from transformers import TFBertModel
from sklearn.preprocessing import LabelEncoder
```

```python
from tensorflow.keras.models import load_model
from transformers import TFBertModel

# Load the model with custom objects
model = load_model(
    '/kaggle/input/model-fix-2/keras/default/1/cross_val_model.h5',
    custom_objects={'TFBertModel': TFBertModel}
)
```

```python
# Tokenisasi kolom 'prepro_text'
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('indobenchmark/indobert-base-p2')
```

```python
df = pd.read_csv('/kaggle/input/dataframe-full/clean_data.csv')
df.rename(columns={'content': 'text', 'Klasifikasi BERT': 'intent'}, inplace=True)
df = df[['text', 'intent']]
```

```python
# Load LabelEncoder (pastikan LabelEncoder sudah fit dengan data pelatihan)
le = LabelEncoder()
le.fit(df['intent'])  # Gunakan intent dari data pelatihan
res_dict = {index: intent for index, intent in enumerate(le.classes_)}

# Tokenisasi seluruh teks di input_data['prepro_text']
tokens = tokenizer(
    input_data['prepro_text'].tolist(),
    max_length=128,
    padding='max_length',
    truncation=True,
    return_tensors="tf"
)

# Prediksi intent dari model
probabilities = model.predict([tokens['input_ids'], tokens['attention_mask']])  # (batch_size, num_classes)

# Ambil indeks dengan probabilitas tertinggi
max_indices = np.argmax(probabilities, axis=1)
max_confidences = np.max(probabilities, axis=1)

# Konversi indeks ke intent menggunakan res_dict
predicted_intents = [res_dict[idx] for idx in max_indices]

# Tambahkan hasil intent & confidence ke input_data
input_data['intent'] = predicted_intents
input_data['confidence'] = max_confidences

# Tampilkan hasil
pd.set_option('display.max_colwidth', None)
input_data[['prepro_text', 'intent', 'confidence']]
```

```python
input_data.to_csv("predicted_intents.csv", index=False)
```
